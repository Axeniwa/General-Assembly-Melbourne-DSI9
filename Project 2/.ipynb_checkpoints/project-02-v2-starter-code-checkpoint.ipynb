# Question: is 65+ age group using stimulant more frequently than all other age groups combined?

# Deliverables: that 65+ age group is/(is not) using stimulant more frequently than all other age groups combined
...

````python
# Code
sti = drug[['age','n','stimulant-frequency']]
sti.loc[drug['age']!='65+'].set_index('age').sum()
````

````python
sti.loc[drug['age'] == '65+']
````

````python
#from our sample it looks though 65+ group use stimulant far more frequently than all other age group combined
# however, we do not know how much we can be confident about this data, to draw a meaningful conclusion.
# hence we need to do a t test
# t test is on mean, what we have is median, I don't know how to do hypothesis analysis on median so I assume the sample distribution is normal, hence mean = median
# but then I don't even know what the standard deviation for the samples are, so I give up. change question.
````

````python
## Question: is there a difference in average drug use proportion between 17 and 21 year old 
## Deliverables: that there is/isn't a difference in average drug use proportion between 17 and 21 year old 
````

````python
fil = pd.concat([drug.iloc[:,0:2],drug.filter(regex='-use')],axis = 1)
````

````python
f17 = fil.loc[fil['age'] =='17']
````

````python
f21 = fil.loc[fil['age'] =='21']
````

````python
arr17 = f17.to_numpy()[0][2:]
````

````python
arr21 = f21.to_numpy()[0][2:]
````

````python
var17 = arr17.var(ddof = 1)
````

````python
var21 = arr21.var(ddof = 1)
````

````python
s = np.sqrt((var17 + var21)/2)
````

````python
t = (arr17.mean() - arr21.mean())/(s*np.sqrt(2/len(arr17)))
````

````python
p = 1 - stats.t.cdf(t,df=len(arr17)*2-2)
````

````python
p
````

````python
t
````

````python
#recheck with internal function
t2, p2 = stats.ttest_ind(arr17,arr21)
````

````python
p2
````

**Report**


Our P value is much larger than our significance level, 0.05, so we have to accept the null hypothesis, that there is no difference in drug use proportion between 17 and 21 year olds.

Outliers are an interesting problem in statistics, in that there is not an agreed upon best way to define them. Subjectivity in selecting and analyzing data is a problem that will recur throughout the course.

1. Pull out the rate variable from the sat dataset.

5. Compare the mean, median, and standard deviation of the "cleaned" data without outliers to the original. What is different about them and why?

````python
sr = satd['Rate']
````

### 9.2 Percentile scoring

Look up percentile scoring of data. In other words, the conversion of numeric data to their equivalent percentile scores.

http://docs.scipy.org/doc/numpy-dev/reference/generated/numpy.percentile.html

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.percentileofscore.html

1. Convert `Rate` to percentiles in the sat scores as a new column.
2. Show the percentile of California in `Rate`.
3. How is percentile related to the spearman rank correlation?

````python
satd['rper'] = satd['Rate'].apply(lambda x:stats.percentileofscore(satd.Rate,x))
````

````python
satd.loc[satd['State'] == 'CA']['rper']
````

````python
#3A: to obtain percentile one needs to reversely rank the variables, divide by the total number of variables and multiply by 100
#3A: hence, percentile is in essence the same as rank, except that it reflects the total number, which rank does not
#3A: with percentile, one is able to derive the rank and be able to calculate the spearman rank correlation
````
